```python
import random
filename = "NegativeSample_casestudy.csv"                         # Read the file name and remove the  \ and / to the left of the file name.
seeds = [666]
with open(filename, "r", encoding="utf-8") as f:
    lines = f.readlines()
lines = [line.strip(" \n") for line in lines]                    # Remove newlines and Spaces
lines = [line for line in lines if len(line)>0]                  # Remove blank lines
errLines = [line for line in lines if len(line.split(","))!=2]
assert len(errLines)==0, "格式错误：\n" + "\n".join(errLines)    # Check the format, whether each row is two columns

n = len(lines)
for seed in seeds:
    random.seed(0)
    indices = [i for i in range(n)]
    for i in range(n-1, -1, -1):                                 # The Gaunard algorithm creates the scrambled index
        idx = int(random.random() * (i + 1))
        indices[i], indices[idx] = indices[idx], indices[i]
    lines_shuffled = [lines[idx] for idx in indices]
    with open("sf_%d_%s" % (seed, filename), "w", encoding="utf-8") as f:
        f.write("\n".join(lines_shuffled))
```
